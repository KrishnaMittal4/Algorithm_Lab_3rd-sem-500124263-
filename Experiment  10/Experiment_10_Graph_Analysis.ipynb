```python
import matplotlib.pyplot as plt  
import numpy as np  
```

```python
tests = ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5']  
naive_times = [0.003403, 0.000869, 0.001319, 0.000740, 0.000759]  
rabin_karp_times = [0.014194, 0.009250, 0.005341, 0.006707, 0.006349]  
kmp_times = [0.008410, 0.008083, 0.008049, 0.009047, 0.007934]  

 
bar_width = 0.25  


x = np.arange(len(tests))  

  
plt.bar(x - bar_width, naive_times, width=bar_width, label='Naive String Matching', color='blue')  
plt.bar(x, rabin_karp_times, width=bar_width, label='Rabin-Karp Algorithm', color='orange')  
plt.bar(x + bar_width, kmp_times, width=bar_width, label='KMP Algorithm', color='green')  


plt.xlabel('Tests')  
plt.ylabel('Time (seconds)')  
plt.title('Execution Time Comparison of String Matching Algorithms')  
plt.xticks(x, tests)  
plt.legend()  


plt.tight_layout()  
plt.show()
```
